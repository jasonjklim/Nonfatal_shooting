```python
# Filter just 2018 data and pop_total column

cities_2018_df = cities_df[cities_df['YEAR']==2018]
```

```python
cities_2018_df['ST_NAME'].str[1:].head()
```

```python
# city name includes 'city' at the end and there is a space in front of st name

# Excldue city
cities_2018_df["CITY_NAME"] = cities_2018_df["CITY_NAME"].str.replace(' city','')
# Excldue space
cities_2018_df['ST_NAME']=cities_2018_df['ST_NAME'].str[1:]
```

```python
cities_2018_df['CITY_NAME'].loc[cities_2018_df['CITY_NAME']=='New York'] = 'NYC'

cities_2018_df['CITY_NAME'].loc[cities_2018_df['CITY_NAME']=='St. Louis'] = 'Saint Louis'

cities_2018_df['CITY_NAME'].loc[cities_2018_df['CITY_NAME']=='Indianapolis (balance)'] ='Indianapolis'

cities_2018_df['CITY_NAME'].loc[cities_2018_df['CITY_NAME']=='Louisville/Jefferson County metro government (balance)'] = 'Louisville'

cities_2018_df['CITY_NAME'].loc[cities_2018_df['CITY_NAME']=='Nashville-Davidson metropolitan government (balance)'] = 'Nashville'
```

```python
df_19N20_3 = df_19N20_2.merge(cities_2018_df, left_on=['City Or County','State'], right_on=['CITY_NAME','ST_NAME'])
```

```python
df_19N20_3 = df_19N20_3[['City Or County', 'State', 'Total Fatalities_2020',
       'Total Non-Fatal Injuries_2020', 'Total Incidents_2020',
       'Total injured or killed_2020', 'Total Fatalities_2019',
       'Total Non-Fatal Injuries_2019', 'Total Incidents_2019',
       'Total injured or killed_2019','POP_TOTAL']]
```

```python
df_19N20_3.head()
```

```python
df_19N20_3.info()
```

```python
# annual rates of the month
df_19N20_3['One_Month_Population'] = round(df_19N20_3['POP_TOTAL']/12, 0)
```

```python
## Creating Columns for Differences

df_19N20_3['Diff_Fatalities'] = df_19N20_3['Total Fatalities_2020'] - df_19N20_3['Total Fatalities_2019']
df_19N20_3['Diff_NonFatal_Injuries'] = df_19N20_3['Total Non-Fatal Injuries_2020'] - df_19N20_3['Total Non-Fatal Injuries_2019']
df_19N20_3['Diff_Incidents'] = df_19N20_3['Total Incidents_2020'] - df_19N20_3['Total Incidents_2019']
```

```python
# rates per 100k population for 2019 and 2020 fatal, non-fatal, incidents

df_19N20_3['Fatality_2020_per_100k'] = round(100000*(df_19N20_3['Total Fatalities_2020']/df_19N20_3['One_Month_Population']),2)
df_19N20_3['NonFatal_2020_per_100k'] = round(100000*(df_19N20_3['Total Non-Fatal Injuries_2020']/df_19N20_3['One_Month_Population']),2)
df_19N20_3['Incidents_2020_per_100k'] = round(100000*(df_19N20_3['Total Incidents_2020']/df_19N20_3['One_Month_Population']),2)


df_19N20_3['Fatality_2019_per_100k'] = round(100000*(df_19N20_3['Total Fatalities_2019']/df_19N20_3['One_Month_Population']),2)
df_19N20_3['NonFatal_2019_per_100k'] = round(100000*(df_19N20_3['Total Non-Fatal Injuries_2019']/df_19N20_3['One_Month_Population']),2)
df_19N20_3['Incidents_2019_per_100k'] = round(100000*(df_19N20_3['Total Incidents_2019']/df_19N20_3['One_Month_Population']),2)
```

```python
df_19N20_3.head()
```

```python
# % difference in rates for fatal, non-fatal, incidents

df_19N20_3['Per_Diff_Fatal_per_100K'] = 100*round((df_19N20_3['Fatality_2020_per_100k'] - df_19N20_3['Fatality_2019_per_100k'])
                                             /df_19N20_3['Fatality_2019_per_100k'], 4)


df_19N20_3['Per_Diff_NonFatal_per_100K'] = 100*round((df_19N20_3['NonFatal_2020_per_100k'] - df_19N20_3['NonFatal_2019_per_100k'])
                                             /df_19N20_3['NonFatal_2019_per_100k'], 4)

df_19N20_3['Per_Diff_Incidents_per_100K'] = 100*round((df_19N20_3['Incidents_2020_per_100k'] - df_19N20_3['Incidents_2019_per_100k'])
                                             /df_19N20_3['Incidents_2019_per_100k'], 4)
```
