```python
import pandas as pd
import numpy as np
import glob

import datetime
datetime.datetime.strptime

pd.set_option('display.max_rows', None)
```

```python


path = r'/Users/jasonlim/Downloads/Non_fatal' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame.head()
```

```python
frame.sort_values(by='Incident Date', inplace=True)
```

```python
frame.tail()
```

```python
frame.info()
```

```python
new_df = frame[['Incident ID', 'Incident Date', 'State', 'City Or County','# Killed', '# Injured']]
```

```python
# new_df.to_csv("non_fatal.csv", index=False)
```

```python
new_df.groupby(['City Or County','State']).sum().sort_values('# Injured', ascending=False)
```

```python
# Top 50

df_2020 = new_df.groupby(['City Or County','State']).agg({'# Killed':'sum',
                                                '# Injured':'sum',
                                                'Incident ID':'count'}).sort_values('# Injured', ascending=False).head(50)
```

```python
df_2020.reset_index(inplace=True)
```

```python
# Pull 2019 file

df_2019_all = pd.read_csv(r"/Users/jasonlim/Documents/GitHub/GVA/All shootings/2019_GVA_shootings.csv")
```

```python
df_2019_all.head()
```

```python
df_2019_all.info()
```

```python
df_2019_all['date'] = pd.to_datetime(df_2019_all['date'])
```

```python
# 30 days from memorial Weekend of 2019 is 5/25/2019-6/24/2019

memo_2019 = df_2019_all[(df_2019_all['date']>='2019-05-25') & (df_2019_all['date']<='2019-06-24')] 
```

```python
memo_2019.head()
```

```python
memo_2019.columns
```

```python
# Group by same method
memo_2019_group = memo_2019.groupby(['city','state']).agg({'killed':'sum',
                                                'injured':'sum',
                                                'id':'count'}).sort_values('id', ascending=False).reset_index()
```

```python
memo_2019_group.head()
```

```python
# combine two dfs, main is 2020. 

df_19N20 = df_2020.merge(memo_2019_group, left_on=['City Or County','State'], right_on=['city','state'])
```

```python
df_19N20
```

```python
df_19N20.columns
```

```python
df_19N20.columns = ['City Or County', 'State', 'Total Fatalities_2020', 'Total Non-Fatal Injuries_2020', 'Total Incidents_2020',
       'city', 'state', 'Total Fatalities_2019', 'Total Non-Fatal Injuries_2019', 'Total Incidents_2019']
```

```python
# Filter out only needful columns

df_19N20_2= df_19N20[['City Or County', 'State', 'Total Fatalities_2020', 'Total Non-Fatal Injuries_2020', 'Total Incidents_2020',
          'Total Fatalities_2019', 'Total Non-Fatal Injuries_2019', 'Total Incidents_2019']]
```
